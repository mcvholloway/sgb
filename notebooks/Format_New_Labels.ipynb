```python
import pandas as pd
```

```python
labels = pd.read_excel("../data/new_2024/Supragastric Belching AI Labels BR Coded Nov30.xlsx")
```

```python
labels
```

```python
labels = labels[labels.index != 29]
```

```python
labels = labels.rename(columns = {'Unnamed: 4': 'patient', 'Events': 'labels'})
labels['patient'] = labels['patient'].astype(int)
```

```python
for study_num in labels['patient'].unique():
    labels[labels['patient'] == study_num][['patient', 'labels']].to_csv(f"../data/new_2024/formatted/labels_{study_num}.csv", index=False)
```
